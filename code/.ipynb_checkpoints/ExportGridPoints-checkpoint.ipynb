```python
%matplotlib notebook
import netCDF4 as nc
import numpy as np
import pandas as pd
from helpers import writebathy, expandf
from helpers import gete1, gete2, writecoords, t2u, t2v, t2f
import matplotlib.pyplot as plt
import scipy.io as sio

#For a full grid - coordfile
#for just water - bathy
coordfilename = '..//data//grid//coordinates_salishsea_1500m.nc'
coordfilename_ORAS5 = '..//data//reference//ORAS5 Mask and Bathy//mesh_mask.nc'
bathyfilename = '..//data//bathymetry//bathy_salishsea_1500m_20210406.nc'

def loadcoords(f): 
    with nc.Dataset(coordfilename) as ncid:
        lat_t = ncid.variables["gphit"][0, :, :].filled()
        lon_t = ncid.variables["glamt"][0, :, :].filled()
    return lat_t, lon_t


def loadbathy(f):
    with nc.Dataset(f) as ncid:
        depth = ncid.variables["Bathymetry"][:, :].filled()
        latt = ncid.variables["nav_lat"][:, :].filled()
        lont = ncid.variables["nav_lon"][:, :].filled()
        
    return depth, latt, lont
    
# get coords
depth, lat_t, lon_t = loadbathy(bathyfilename)
lat_t2, lon_t2 = loadcoords(coordfilename_ORAS5)


i = 0 
coords = []
while i < len(lat_t[:,0]):
    
    coords_row = [list(coord) for coord in zip(lat_t[i,:], 
                                               lon_t[i,:], 
                                               depth[i,:], 
                                               np.arange(0,lat_t.shape[1]), 
                                               np.full((lat_t.shape[1]), i))]
    coords = coords + coords_row

    i += 1
    
df = pd.DataFrame(coords)


i = 0 
coords2 = []
while i < len(lat_t2[:,0]):
    
    coords_row = [list(coord) for coord in zip(lat_t2[i,:], 
                                               lon_t2[i,:],
                                               np.arange(0,lat_t2.shape[1]), 
                                               np.full((lat_t2.shape[1]), i))]
    coords2 = coords2 + coords_row

    i += 1
    
df2 = pd.DataFrame(coords2)
```
